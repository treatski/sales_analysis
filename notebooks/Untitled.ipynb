```python
import pandas as pd
import sys
sys.path.insert(1, '../src')
import treats_work
```

```python
contract, cust = treats_work.load_data()
contract.head()
```

```python
cust.head()
```

```python
df = treats_work.combined_data(contract, cust)
df.head ()
```

```python
df['deposit'] = pd.to_numeric(df['deposit'].str.replace("[$,]",""))

df['monthly_amt'] = pd.to_numeric(df['monthly_amt'].str.replace("[$,]",""))
df['total']= df['deposit'] + (df['monthly_amt']* df['contract_term'])
```

```python
df.head()
```

```python
dfstate= df.groupby("state").agg({'company':pd.Series.nunique, 'total':"sum"})
dfstate.head()
```

```python
dfstate=dfstate.sort_values("total", ascending=False)
dfstate
```
